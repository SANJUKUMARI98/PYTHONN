```python
data2=pd.DataFrame(data2)
```

```python
pd.merge(data1,data2)  # based on common values intersection
# key1 is common in both and matching data and place it otherwise nan
```

```python
pd.merge(data1,data2,how='left')
# calling everything from left  and wherever is a match it place it 
# else nan
```

```python
pd.merge(data1,data2,how='right')
# taking all data from right and wherever it is matching it place it or else nan
```

```python
pd.merge(data1,data2,how='outer') # joinig both data1,data2 
```

```python
pd.merge(data1,data2,how='cross') #combining all the combination
```

```python
pd.merge(data1,data2,how = 'left',on ='key1')
```

```python
pd.merge(data1,data2,how='left',left_on='key2',right_on='key4')
taking left join  checking value 
```
